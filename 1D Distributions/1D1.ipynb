```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb 
import pandas as pd
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
d1 = np.loadtxt("example_1.txt")
d2 = np.loadtxt("example_2.txt")
print (d1.shape, d2.shape)
```

## Histogram Plots

```python
plt.hist(d1, label="D1")
plt.hist(d2, label="D2")
plt.legend()
plt.ylabel("Counts")
```

```python
counts1m, bins, _ = plt.hist(d2, bins = 10, label="D2")
plt.hist(d1, bins=bins, label="D1")
plt.legend()
plt.ylabel("Counts");
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist(d2, bins=bins, label="D2")
plt.hist(d1, bins=bins, label="D1")
plt.legend()
plt.ylabel("Counts");
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist(d2, bins=bins, label="D2", density=True)
plt.hist(d1, bins=bins, label="D1", density=True)
plt.legend()
plt.ylabel("Probability");
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist(d2, bins=bins, label="D2", density=True, histtype='step', ls=":", lw=2)
plt.hist(d1, bins=bins, label="D1", density=True, histtype='step', lw=3)
plt.legend()
plt.ylabel("Probability");
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist(d2, bins=bins, label="D2", density=True, histtype='barstacked', ls=":", lw=2)
plt.hist(d1, bins=bins, label="D1", density=True, histtype='barstacked', lw=3, alpha=0.5)
plt.legend()
plt.ylabel("Probability");
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist([d1,d2], bins=bins, label="Stacked", density=True, histtype='barstacked', ls=":", lw=2)
plt.legend()
plt.ylabel("Probability");
```

```python
bins = np.linspace(min(d1.min(),d2.min()),max(d1.max(),d2.max()),50)
plt.hist([d1,d2], bins=bins, label="Stacked", density=True, histtype='barstacked', ls=":", lw=2)
plt.hist(d2, bins=bins, label="D2", density=True, histtype='step', ls=":", lw=2)
plt.hist(d1, bins=bins, label="D1", density=True, histtype='step', lw=3, alpha=0.5)

plt.legend()
plt.ylabel("Probability");
```

```python
bins = 50
plt.hist([d1,d2], bins=bins, label="Stacked", density=True, histtype='barstacked', ls=":", lw=2)
plt.hist(d2, bins=bins, label="D2", density=True, histtype='step', ls=":", lw=2)
plt.hist(d1, bins=bins, label="D1", density=True, histtype='step', lw=3, alpha=0.5)

plt.legend()
plt.ylabel("Probability");
```

### Bee Swarm Plots

```python
dataset = pd.DataFrame({
    'value':np.concatenate((d1, d2)),
    'type':np.concatenate((np.ones(d1.shape),np.zeros(d2.shape)))
})
dataset.info()
```

```python
sb.swarmplot(dataset["value"])
```

```python
sb.swarmplot(x="type", y="value", data=dataset, size=2)
```

### Box plots

```python
sb.boxplot(x='type', y='value', data=dataset)
sb.swarmplot(x='type', y='value', data=dataset, size=2, color="k", alpha=0.5)
```

### Empirical Cumulative Distribution Function

```python
cdf = np.linspace(1/d1.size, 1, d1.size)
sd1 = np.sort(d1)
sd2 = np.sort(d2)

plt.plot(sd1, cdf, label = "DS1 CDF")
plt.plot(sd2, cdf, label = "DS2 CDF")

plt.hist(d1, histtype='step', density=True, alpha=0.7)
plt.hist(d2, histtype='step', density=True, alpha=0.7)
plt.legend()
```
