# Preparing Dataset

```python
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
df = df.fillna(0)
```

```python
df1 = df[['Glucose','BMI','Age','Outcome']]
df1.head()
```

```python
df1.describe()
```

```python
df2 = df1.loc[~(df1[df1.columns[:-1]] == 0).any(axis=1)]
df2.head()
```

```python
df2.describe()
```

```python
df2.groupby('Outcome').mean()
```

```python
df2.groupby('Outcome').agg(['mean','median'])
```

```python
df2.groupby('Outcome').agg({'Glucose':'mean','BMI':'median','Age':'median'})
```

```python
positive = df2.loc[df2['Outcome']==1]
negative = df2.loc[df2['Outcome']==0]
```

```python
print(positive.shape, negative.shape)
```

```python
df2.to_csv('clean_diabetes_my', index=False)
```
